```python
import pandas as pd
import numpy as np
```

```python
rng = pd.date_range('1/1/2011', periods=72, freq='H')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

```python
converted = ts.asfreq('45Min', method='pad')
converted
```

```python
converted = ts.asfreq('90Min', method = 'bfill')
converted
```
